```python
#Tasck 3
mean_exp=df['WorkExp'].mean()
median_exp=df['WorkExp'].median()
mode_exp=df['WorkExp'].mode()[0]

result = pd.DataFrame({
    'Metric': ['Mean', 'Median', 'Mode'],
    'Value': [mean_exp, median_exp, mode_exp]
})
result=result.round(2)
print(result)
```

```python
#Tasck7
python_df = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
python_df = python_df[python_df['ConvertedCompYearly'].notna()]
result = python_df.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).reset_index()
result = result.rename(columns={
    'mean': 'AverageCompensation',
    'median': 'MedianCompensation'
})

print(result)
```

```python
#Tasck 8
salary = df[df['ConvertedCompYearly'].notna()]
sort = salary.sort_values(by='ConvertedCompYearly', ascending=False)
top5 = sort.head(5)
result = top5[['EdLevel', 'ConvertedCompYearly']]
print(result)
```

```python
#Tasck 9
df['Age'].unique()
```

```python
#Создадим флаг Python-разработчика
df['is_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

result = df.groupby('Age')['is_python'].mean().reset_index()
result['PythonPercent'] = result['is_python'] * 100
result = result[['Age', 'PythonPercent']]
result = result.round(2)
print(result)
```

```python
#Tasck 10
salary = df[df['ConvertedCompYearly'].notna()]
q75 = salary['ConvertedCompYearly'].quantile(0.75)
filtered = salary[(salary['ConvertedCompYearly'] > q75) &(salary['RemoteWork'] == 'Remote')]
industry_counts = filtered['Industry'].value_counts()

print(industry_counts)
```
